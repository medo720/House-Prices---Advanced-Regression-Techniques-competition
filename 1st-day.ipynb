```python
train_df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
test_df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
```

```python
train_df.drop(['PoolQC','Fence','MiscFeature','Alley','FireplaceQu'],axis = 'columns',inplace = True)
test_df.drop(['PoolQC','Fence','MiscFeature','Alley','FireplaceQu'],axis = 'columns',inplace = True)
```

```python
num_cols = train_df.select_dtypes(include=np.number).columns.tolist()
cat_cols = train_df.select_dtypes(exclude=np.number).columns.tolist()
print(len(num_cols))
print(len(cat_cols))
```

```python
miss_num_cols = train_df.loc[:,num_cols].columns[train_df.loc[:,num_cols].isnull().any()].tolist()
miss_cat_cols = train_df.loc[:,cat_cols].columns[train_df.loc[:,cat_cols].isnull().any()].tolist()
```

```python
print(miss_num_cols)
print(miss_cat_cols)
```

```python
train_df[['MasVnrArea']]
```

```python
imp_num = SimpleImputer(strategy = 'mean')
imp_cat = SimpleImputer(strategy = 'most_frequent')
train_df[miss_num_cols] = imp_num.fit_transform(train_df[miss_num_cols])
train_df[miss_cat_cols] = imp_cat.fit_transform(train_df[miss_cat_cols])
test_df[miss_num_cols] = imp_num.fit_transform(test_df[miss_num_cols])
test_df[miss_cat_cols] = imp_cat.fit_transform(test_df[miss_cat_cols])


```

```python
train_df.info()
```

```python
profile = ProfileReport(train_df,title = 'Pandas Profiling Report"',explorative = True)
```
